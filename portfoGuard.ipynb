# Casovy vyvoj

```python
def get_units(ticker: str, date_to: str) -> float:
    return trades[(trades.ticker == ticker) & (trades.date < date_to)]["units"].sum()


def get_total_price(ticker: str, date_to: str) -> float:
    return trades[(trades.ticker == ticker) & (trades.date < date_to)]["total_price"].sum()
```

```python
# Aktuln datum
current_date = datetime.now().date()
# Vytvoen seznamu poslednch ticeti dn
last_thirty_days = [
    (current_date - timedelta(days=i)).strftime("%Y-%m-%d") for i in range(n_days)
]
safety_range = (current_date - timedelta(days=n_days + 10)).strftime("%Y-%m-%d")
last_thirty_days = sorted(last_thirty_days)
time_evol = pd.DataFrame(columns=["date", "ticker", "value","total_price"])
for ticker in summary["ticker"].unique():
    # for ticker in ["CEZ.PR"]:
    currency = summary[summary["ticker"] == ticker].iloc[0]["currency"]
    stock = yf.Ticker(ticker)
    history = stock.history(start=safety_range, end=last_thirty_days[-1])[
        "Close"
    ].reset_index()
    history["Date"] = history["Date"].dt.strftime("%Y-%m-%d")
    # print(history)
    for date in last_thirty_days:
        # print(date)
        if date in history["Date"].values:
            price = float(history[history["Date"] == date]["Close"])
            # print("Found:",price)
        else:
            price = history[history["Date"] < date].iloc[-1]["Close"]
            # print("Not found",price)
        time_evol.loc[len(time_evol)] = [date,ticker,price * get_exchange_rate(currency, date) * get_units(ticker, date), get_total_price(ticker,date)]
        # print(index, price, get_exchange_rate(currency, index))
# print(time_evol)
time_evol["gain"] = time_evol["value"]/time_evol["total_price"]*100-100
```

```python
watchlist = read_watchlist()
watchlist
```

```python
# Aktuln datum
current_date = datetime.now().date()
n_days = 60
start_date = current_date - timedelta(days=n_days)
# Vytvoen seznamu poslednch ticeti dn
time_evol = pd.DataFrame(columns=["date", "ticker", "ratio"])
for ticker in watchlist["ticker"]:
    currency = watchlist[watchlist["ticker"] == ticker].iloc[0]["currency"]
    price_limit = watchlist[watchlist["ticker"] == ticker].iloc[0]["price_limit"]
    stock = yf.Ticker(ticker)
    stock_history = stock.history(
        start=start_date.strftime("%Y-%m-%d"), end=current_date.strftime("%Y-%m-%d")
    )["Close"].reset_index()
    stock_history["Date"] = stock_history["Date"].dt.strftime("%Y-%m-%d")
    stock_history["ratio"] = stock_history["Close"] / price_limit
    for id,row in stock_history.iterrows():
        time_evol.loc[len(time_evol)] = [
            row["Date"],
            ticker,
            row["ratio"]
        ]
```

```python
fig = px.line(
    time_evol,
    x="date",
    y="ratio",
    color="ticker",
    title="Buy ratio",
    template="plotly_dark",
    width=1200,
    height=800,
)

# showing the plot
fig.show()
```
